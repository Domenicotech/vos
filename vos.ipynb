```python
import json
def load_json(file):
    '''
    The function loads a json file
    input:
        file: json file pathù
    
    It returns
        the json object
    '''
    with open(file, 'r') as f:
        a = json.load(f)
    return a

def save_json(obj, file, indent=4, sort_keys=True):
    '''
    The function saves a json file
    input:
        obj: an object to save
        file: the saving file path
        indent: int, the indent spacing
        sort_keys: bool, True if you want the keys sorted
    
    '''
    with open(file, 'w') as f:
        json.dump(obj, f, sort_keys=sort_keys, indent=indent)
```

```python
label= load_json('/content/root/fiftyone/coco-2017/validation/val_coco_format.json')
```

```python
new = load_json('/content/VOC_0712_converted/val_coco_format.json')
```

```python
#individua i nomi delle categorie che ci interessano
used_names = [cat['name'] for cat in new['categories']]
```

```python
#filtra le categorie che ci interessano
categories = [cat for cat in label['categories'] if cat['name'] in used_names]
```

```python
#filtra le annotazioni che ci interessano (in base alle categorie rimaste)
ids = [cat['id'] for cat in categories]
annotations = [ann for ann in label['annotations'] if ann['category_id'] in ids]
```

```python
#rimappa gli id delle categorie che ci interessano, in base alla vecchia nomenclatura


#new_id: old_id mapping
remapping_dic = {}


def remap(category):
  new_id = [cat['id'] for cat in new['categories'] if cat['name']==category['name']]
  remapping_dic[category['id']] = new_id[0]
  category['id'] = new_id[0]
  return category
categories = [remap(cat) for cat in categories]
```

```python
def remap_ann(annotation):
  annotation['category_id'] = remapping_dic[annotation['category_id']]
  return annotation
annotations = [remap_ann(ann) for ann in annotations]
```

```python
label['categories']= categories
label['annotations'] = annotations
save_json(label, '/content/root/fiftyone/coco-2017/validation/val_coco_format.json')
```
